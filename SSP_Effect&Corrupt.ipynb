```python
dt = df[df['Variable'] == 'Governance Index|Government Effectiveness']
dt_OB = df_OB[df_OB['Variable'] == 'Governance Index|Government Effectiveness']
dt_OB = dt_OB.drop(columns =  ["Model","Scenario","Variable"])
dt_OB = dt_OB.set_index('Region')
dt_OB = pd.DataFrame(dt_OB.stack(dropna = False), 
                    columns = ['val']).reset_index().rename(columns={'level_1':'year'})
dt_OB['year'] = dt_OB['year'].apply(pd.to_numeric, errors='coerce')

for i in range(1,6):
    dd = dt[dt['Scenario'] == f'SSP{i}']
    print(i)
    dd = dd.drop(columns =  ["Model","Scenario","Variable"])
    dd = dd.set_index('Region')
    dd = pd.DataFrame(dd.stack(dropna = False), 
                    columns = ['val']).reset_index()#.rename(columns={'level_1':'year'})
    dd['level_1'] = dd['level_1'].apply(pd.to_numeric, errors='coerce')
    dd['year'] = dd['level_1'] + 1
    dd = dd[['Region','year','val']]
    dd = pd.concat([dt_OB,dd], ignore_index= True)
    dd.val = round(dd.val*5, 4)
    first_last = dd.sort_values(['Region','year']).groupby(dd['Region'])['val'] \
            .agg(['first', 'last']).reset_index()
    first_last = first_last.rename(columns={"first": "Earliest", "last": "MostRecent"})
    dd = dd.pivot_table(index = 'Region',
                        columns='year',
                        values = 'val').reset_index()
    dd = dd.merge(first_last, on = 'Region', how = 'left')
    dd = country.merge(dd, on = 'Region', how = 'left')
    dd = dd.drop(columns=['Region'])
#     # dd.columns = col
    # dd.to_csv(f'IFs Import/ForecastGovernanceEffectSSP{i}Andrijevic.csv', index = False)
dd
```

## Governance Index|Control of Corruption

```python
dt = df[df['Variable'] == 'Governance Index|Control of Corruption']
dt_OB = df_OB[df_OB['Variable'] == 'Governance Index|Control of Corruption']
dt_OB = dt_OB.drop(columns =  ["Model","Scenario","Variable"])
dt_OB = dt_OB.set_index('Region')
dt_OB = pd.DataFrame(dt_OB.stack(dropna = False), 
                    columns = ['val']).reset_index().rename(columns={'level_1':'year'})
dt_OB['year'] = dt_OB['year'].apply(pd.to_numeric, errors='coerce')

for i in range(1,6):
    dd = dt[dt['Scenario'] == f'SSP{i}']
    print(i)
    dd = dd.drop(columns =  ["Model","Scenario","Variable"])
    dd = dd.set_index('Region')
    dd = pd.DataFrame(dd.stack(dropna = False), 
                    columns = ['val']).reset_index()
    dd['level_1'] = dd['level_1'].apply(pd.to_numeric, errors='coerce')
    dd['year'] = dd['level_1'] + 1
    dd = dd[['Region','year','val']]
    dd = pd.concat([dt_OB,dd], ignore_index= True)
    dd.val = round(dd.val*10, 4)
    first_last = dd.sort_values(['Region','year']).groupby(dd['Region'])['val'] \
            .agg(['first', 'last']).reset_index()
    first_last = first_last.rename(columns={"first": "Earliest", "last": "MostRecent"})
    dd = dd.pivot_table(index = 'Region',
                        columns='year',
                        values = 'val').reset_index()
    dd = dd.merge(first_last, on = 'Region', how = 'left')
    dd = country.merge(dd, on = 'Region', how = 'left')
    dd = dd.drop(columns=['Region'])
#     dd.columns = col
    dd.to_csv(f'IFs Import/ForecastGovernanceCorruptionSSP{i}Andrijevic.csv', index = False)
```
